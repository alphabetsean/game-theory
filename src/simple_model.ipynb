```python
from scipy.stats import truncnorm
from scipy.stats import bernoulli
import numpy as np
import matplotlib.pyplot as plt
```

```python
Z = 500
W = 9000
F = 1500
D = -1000
L = -30
P = 400
sigma1 = 0.1
sigma2 = 0.1
```

```python
day = 150
N_s = np.zeros(day)
N_i = np.zeros(day)
O_s = np.zeros(day)
O_i = np.zeros(day)
N_s[0] = 9700
N_i[0] = 300
```

```python
for i in range(day):
    if i == 0:
        mu1 = 0.5
        mu2 = 0.5
    else:
        mu1 = O_s[i-1] / N_s[i-1]
        mu2 = O_i[i-1] / N_i[i-1]
    clip_a = 0.0 
    clip_b = 1.0
    
    a1, b1 = (clip_a - mu1) / sigma1, (clip_b - mu1) / sigma1
    a2, b2 = (clip_a - mu2) / sigma2, (clip_b - mu2) / sigma2
    Splayer_estiS = truncnorm.rvs(a1, b1, loc=mu1, scale=sigma1, size=int(N_s[i])) * N_s[i]
    Splayer_estiI = truncnorm.rvs(a2, b2, loc=mu2, scale=sigma2, size=int(N_s[i])) * N_i[i]
    
    Iplayer_estiS = truncnorm.rvs(a1, b1, loc=mu1, scale=sigma1, size=int(N_i[i])) * N_s[i]
    Iplayer_estiI = truncnorm.rvs(a2, b2, loc=mu2, scale=sigma2, size=int(N_i[i])) * N_i[i]
    Sutil1 = 0 + D * np.clip(Splayer_estiI/Z,0,1) + (1-np.clip((Splayer_estiI+Splayer_estiS-1)/W,0,1))*P
    S_action = Sutil1 > L
    O_s[i] = np.sum(S_action)
    
    Iutil1 = D + (-F*np.clip(Iplayer_estiS/Z,0,1))+ (1-np.clip((Iplayer_estiI+Iplayer_estiS-1)/W,0,1))*P
    I_action = Iutil1 > (D+L)
    O_i[i] = np.sum(I_action)
    print(O_i[i])
    if i == day - 1:
        break
    infected = 0
    for j in range(int(N_s[i])):
        if S_action[j]:
            r = bernoulli.rvs(max(min(O_i[i]/Z, 1), 0),size=1)
            infected += r
    N_s[i+1] = N_s[i] - infected
    N_i[i+1] = N_i[i] + infected
        
        
```
